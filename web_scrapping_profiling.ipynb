```python
import requests
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport 
import matplotlib.pyplot as plt
```

```python
# get table from website using pandas

web_adress = 'https://www.worldometers.info/world-population/population-by-country/'

# Use requests to retrieve data from a given web_adress
website = requests.get(web_adress)
```

```python
# read table from website content
df_list = pd.read_html(website.content)

# read_html returns a list of tables, we need to select one of them
df = df_list[0]

df.head()
```

```python

# Generate the report. 
# We use the title parameter for naming our report and
# explorative parameter True for deeper exploration

profile = ProfileReport(df, title='report', explorative=True)
profile
```
